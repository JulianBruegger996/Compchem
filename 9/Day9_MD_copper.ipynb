# Molecular Dynamics of Copper

Import all necessary libraries.

```python
import ase.units as units
from ase.io.trajectory import Trajectory
from ase.md import Langevin
from ase.visualize import view
import matplotlib.pyplot as plt
import numpy as np
import scipy.constants as const

from ase.calculators.emt import EMT
from ase.lattice.cubic import FaceCenteredCubic as LatticeFCC
from ase.md.velocitydistribution import MaxwellBoltzmannDistribution
```

Create a copper crystal lattice using the ase function LatticeFCC(). It takes as arguments:
- Directions: Miller indices. A 3x3 diagonal matrix (1 along the diagonal) is the standard cartesian alignement.
- Symbol: Element that is simulated
- Size: Tuple how often the unit cell will be repeated in x,y,z direction
- pbc: set periodic boundary conditions to True


```python
# Set up initial positions of Cu atoms on Fcc crystal lattice
copper_crystal_300K = LatticeFCC(symbol="Cu", directions=np.eye(3, dtype=int), size=(5, 5, 5), pbc=True)
```

View the copper crystal. If you use the viewer x3d, it may take some time until it shows.

```python
# view the structure
view(copper_crystal_300K, viewer="x3d")
```

Reuse the function from the water MD to run your simulations. As an additional argument the temperature should be included. Additionally, the positions of the atoms should be saved only every 5th simulation step, otherwise the trajectory file will be too large.
As thermostat we will use Langevin and the friction can be set to 0.05 for the equilibration run, and to 0.02 for the production run. The function will begin with an equilibration run, followed by a production run, both have the same length and stepsize. This needs to be done every time the temperature is changed, otherwise the system will not be equilibrated at the new temperature.

```python
def run_simulation_NVT(dt, steps, temp, tag, atoms):
    # copy the atoms object to avoid modifying the original structure
    # attach the EMT calculator to the copied atoms object
    md_atoms = atoms.copy()
    md_atoms.calc = EMT()

    # Set the initial velocities corresponding to T=temp from Maxwell Boltzmannn distribution
    MaxwellBoltzmannDistribution(md_atoms, temperature_K=temp)

    print("Starting equilibration run...")
    # perform equilibration run
    # change the tag for the equilibration run to avoid overwriting the production run log and trajectory files
    # define the Langevin dynamics object with the copied atoms, time step, temperature, friction coefficient = 0.05, and log file name
    dyn = Langevin(md_atoms, dt, temperature_K=temp, friction=0.05, logfile=f'{tag}_equi.log')
    # attach a trajectory object to the dynamics object to save the trajectory every 5 steps (keyword: interval=5) using the tag for the equilibration run   
    # run the dynamics for the specified number of steps
    traj = Trajectory(f'{tag}_equi.traj', mode="w", atoms=md_atoms)
    dyn.attach(traj.write, interval=5)
    dyn.run(steps)

    print("Finished equilibration! Starting production run...")
    # perform production run
    # define the Langevin dynamics object with the copied atoms, time step, temperature, friction coefficient = 0.02, and log file name (tag) for the production run
    dyn = Langevin(md_atoms, dt, temperature_K=temp, friction=0.02, logfile=f'{tag}.log')
    # attach trajectory object to the dynamics object to save the trajectory every 5 steps (keyword: interval=5) using the tag for the production run
    # run the dynamics for the specified number of steps
    traj = Trajectory(f'{tag}.traj', mode="w", atoms=md_atoms)
    dyn.attach(traj.write, interval=5)
    dyn.run(steps)
    print("Finsished!")
```

Define the time step to be 2 fs and the total simulation time 1 ps. Start the simulations for 300 K.

```python
# define the number of steps, the time step, and the temperatures for the simulations
time_step = 2
steps = 1000 // time_step
temperature = 300
# run the simulations for each temperature using the run_simulation_NVT function. Use a tag that depends on the temperature to avoid overwriting log and trajectory files.
run_simulation_NVT(time_step * units.fs, steps, temperature, f"copper_crystal_{temperature}K", copper_crystal_300K)
```

Now, we want to simulate copper at 1500 K. This time, we will set the periodic boundary conditions to False, so that the atoms can move freely in space. Why is this done here? Discuss this in the report.

```python
# create the new atoms object for the 1500 K simulation with periodic boundary conditions set to False
copper_crystal_1500K = LatticeFCC(symbol="Cu", directions=np.eye(3, dtype=int), size=(5, 5, 5), pbc=False)
```

For the md simulation, we again use time steps of 2 fs and a total simulation time of 1 ps.

```python
# run the simulation
time_step = 2
steps = 1000 // time_step
temperature = 1500
run_simulation_NVT(time_step * units.fs, steps, temperature, f"copper_crystal_{temperature}K", copper_crystal_1500K)
```

Plot the energy and temperature vs time. Use np.loadtxt() to read in the log file that was created during the MD run.

```python
# create arrays to store the time, total energy, potential energy, kinetic energy, and temperature from the log file of the simulations using np.loadtxt. Skip the first row of the log file and unpack the columns into separate arrays.
# The arrays could be called: time, etot, epot, ekin, temp (then the plotting function will work without modification)
time, etot, epot, ekin, temp = np.loadtxt(
    'copper_crystal_1500K.log', 
    skiprows=1, 
    unpack=True
)
```

Plot the distance between two atoms. For this, we will read in the trajectory file. Then, we will loop through the frames of the trajectory and calculate the distance between the two atoms for each frame. Finally, we will plot the distance vs time.

Hint: as the interval at which the trajectory was saved is 5, the time array here is different from the time array in the previous plot.

```python
# Load the trajectory using traj = Trajectory(filename)
traj = Trajectory("copper_crystal_300K.traj")

# Define the indices of the two neighbouring atoms. You can visualize the atoms and their indices using ase gui.
index1 = 0
index2 = 1

# Create empty lists to store the distances and times. Define the time step and save interval based on your simulation settings.
distances = []
times = []

# Loop through frames and calculate the distances
for i, atoms in enumerate(traj):
    # get the distance between the two atoms using atoms.get_distance(atom_i, atom_j, mic=True)
    # mic=True ensures we measure the distance across periodic boundaries --> measures the shortest distance between the two atoms, even if they are on opposite sides of the periodic box
    # append the distance to the distances list and the corresponding time to the times list using the time steps and interval at which the trajectory was saved
    distances.append(atoms.get_distance(0, 1, mic=True))
    times.append(5*i*time_step)
```

```python
# plot the distance as a function of time. Label the axes and give the plot a title. You can also add a grid for better visualization.

plt.plot(times, distances)
plt.xlabel("Time [fs]")
plt.ylabel(r"Distance [$\AA$]")
plt.title("Distance of 2 copper atoms at 300K")
plt.grid()
plt.show()
```

Calculate the wave number of the vibration at 300 K and 1500 K.

```python
def get_peaks(data, time_step, stride):
    peaks = []
    times = []
    previous = data[0]
    current = data[0]

    for i, next in enumerate(data):
        if current > previous and current > next:
            peaks.append(current)
            times.append((i-1) * stride * time_step)
        previous = current
        current = next

    return times, peaks
```

```python
# Analyze the plot to determine the period of the vibration. You can do this by identifying the time points corresponding to peaks. Calculate the average time for one full vibration cycle and use it to calculate the frequency and wave number of the vibration. The results should be given in cm^-1.
time_points, peaks = get_peaks(distances, time_step, 5)
print("Found peaks at: ", time_points)

average_time = np.mean([time_points[i+1] - time_points[i] for i in range(len(time_points) - 1)])
print(f"Average time for a full vibration: {average_time:.1f}fs")
wave_number = 33356/average_time
print(f"Wave number: {wave_number:.3f}cm^-1")
```
