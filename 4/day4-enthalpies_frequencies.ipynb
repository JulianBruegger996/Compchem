# Day 4: Reaction enthalpies and harmonic vibrational frequencies

In this exercise we will calculate the reaction enthalpies of two reactions (H2 + CO --> H2CO and H2 + H2CO --> CH3OH) at 298.15 K. Then, we will calculate and visualize the vibrational modes for formaldehyde.

```python
from ase.visualize import view
from ase.build import molecule
#from pymolpro import ASEMolpro
from ase.optimize import BFGS
import matplotlib.pyplot as plt
from ase import Atoms
from ase.vibrations import Vibrations
from ase.thermochemistry import IdealGasThermo
from ase import units
```

Build the molecules H2, CO, H2CO and CH3OH with ASE. You can visualize them for example with 'x3d' to make sure they are correct. 

```python
# Build Molecules
H2 = molecule("H2")
CO = molecule("CO")
H2CO = molecule("H2CO")
CH3OH = molecule("CH3OH")
molecules = [H2, CO, H2CO, CH3OH]

# visualization
view(H2CO, viewer="x3d")
```

### H2

1. set up a calculator and run a geometry optimization with the ansatz 'DF-MP2/def2-TZVPP'. 
2. calculate the vibrations using the ASE Vibrations tool. Proceed just as in the example shown on https://ase-lib.org/ase/vibrations/modes.html#example. In the example one line of code is "vib.summary(log='H2O_EMT_summary.txt')". You can use instead vib.summary() to get the output printed out in this jupyter notebook. You will obtain the vibration modes and the zero-point energy. Visualize the vibrational modes by writing the mode to a file and opening it with ase gui.
3. Use the ASE Thermochemistry-IdealGasThermo module to obtain enthalpy, entropy and free energy components at a set temperature of 298.15K and a pressure of 101325 Pa. Have a look at the example https://ase-lib.org/ase/thermochemistry/thermochemistry.html#example to see how its done. Leave out the keyword 'vib_selection'. The symmetry number for H2 is 2. Make sure that you enter the correct geometry for each molecule (either linear or non linear).

```python
# 1. Optimise the molecule 
H2.calc = ASEMolpro(ansatz='DF-MP2/def2-TZVPP', density_fitting=True)
with BFGS(H2, trajectory='H2.traj') as opt:
    opt.run(fmax=0.0001)
```

```python
# 2. Vibrations
vibH2 = Vibrations(H2, name="H2")
vibH2.run()
vibH2.summary()
vibH2.write_mode(-1)
```

```python
# 3. Thermochemistry
thermo = IdealGasThermo(
    vib_energies=vibH2.get_energies(),
    potentialenergy=H2.get_potential_energy(),
    atoms=H2,
    geometry='linear',
    symmetrynumber=2,
    spin=0,
)
GH2O = thermo.get_gibbs_energy(temperature=298.15, pressure=101325.0)
HH2O = thermo.get_enthalpy(temperature=298.15)
```

Proceed in the same way for all the other molecules. The symmetry numbers are: CO: 1, H2CO: 2, CH3OH: 1.

### CO

```python
# 1. Optimise the molecule 
CO.calc = ASEMolpro(ansatz='DF-MP2/def2-TZVPP', density_fitting=True)
with BFGS(CO, trajectory='CO.traj') as opt:
    opt.run(fmax=0.0001)
```

```python
# 2. Vibrations
vibCO = Vibrations(CO, name="CO")
vibCO.run()
vibCO.summary()
vibCO.write_mode(-1)
```

```python
# 3. Thermochemistry
thermo = IdealGasThermo(
    vib_energies=vibCO.get_energies(),
    potentialenergy=CO.get_potential_energy(),
    atoms=CO,
    geometry='linear',
    symmetrynumber=1,
    spin=0,
)
GCO = thermo.get_gibbs_energy(temperature=298.15, pressure=101325.0)
HCO = thermo.get_enthalpy(temperature=298.15)
```

### H2CO

```python
# 1. Optimise the molecule 
H2CO.calc = ASEMolpro(ansatz='DF-MP2/def2-TZVPP', density_fitting=True)
with BFGS(H2CO, trajectory='H2CO.traj') as opt:
    opt.run(fmax=0.0001)
```

```python
# 2. Vibrations
vibH2CO = Vibrations(H2CO, name="H2CO")
vibH2CO.run()
vibH2CO.summary()
vibH2CO.write_mode(-1)
```

```python
# 3. Thermochemistry
thermo = IdealGasThermo(
    vib_energies=vibH2CO.get_energies(),
    potentialenergy=H2CO.get_potential_energy(),
    atoms=H2CO,
    geometry='nonlinear',
    symmetrynumber=2,
    spin=0,
)
GH2CO = thermo.get_gibbs_energy(temperature=298.15, pressure=101325.0)
HH2CO = thermo.get_enthalpy(temperature=298.15)
```

### CH3OH

```python
# 1. Optimise the molecule 
CH3OH.calc = ASEMolpro(ansatz='DF-MP2/def2-TZVPP', density_fitting=True)
with BFGS(CH3OH, trajectory='CH3OH.traj') as opt:
    opt.run(fmax=0.0001)
```

```python
# 2. Vibrations
vibCH3OH = Vibrations(CH3OH, name="CH3OH")
vibCH3OH.run()
vibCH3OH.summary()
vibCH3OH.write_mode(-1)
```

```python
# 3. Thermochemistry
thermo = IdealGasThermo(
    vib_energies=vibCH3OH.get_energies(),
    potentialenergy=CH3OH.get_potential_energy(),
    atoms=CH3OH,
    geometry='nonlinear',
    symmetrynumber=1,
    spin=0,
    ignore_imag_modes=True
)
GCH3OH = thermo.get_gibbs_energy(temperature=298.15, pressure=101325.0)
HCH3OH = thermo.get_enthalpy(temperature=298.15)
```

Now you know all the individual terms you need for calculating the reaction enthalpy of both reactions. Don't forget to convert to kJ/mol.  

```python
import numpy as np

gibbs_energies = np.array([GH2O, GCO, GH2CO, GCH3OH])
gibbs_energies *= units.mol/units.kJ
enthalpies = np.array([HH2O, HCO, HH2CO, HCH3OH])
enthalpies *= units.mol/units.kJ

# H2 + CO --> H2CO 
print("Reaction enthalpy of H2 + CO --> H2CO: ", round(enthalpies[2] - enthalpies[1] - enthalpies[0], 2))

# H2 + H2CO --> CH3OH
print("Reaction enthalpy of H2 + H2CO --> CH3OH: ", round(enthalpies[3] - enthalpies[2] - enthalpies[0], 2))
```
